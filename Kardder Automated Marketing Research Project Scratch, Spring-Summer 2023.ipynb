```python
# notes:#tup anatomy t = ("org. name", avg # comments per post, avg # likes per post, # posts, # following #followers)
'''
I am writing a script that will take in an organization's instagram tag and generate a series of useful statistics for
ranking possible/prospective marketing partners (i.e., automated marketing research along a specific platform, Instagram!)


This class fice will store each orgs. profile nicely for evaluation/comparison purposes. Eventually, with enough data, 
it would be cool to train a model to predict something useful from any number of these 5* or so features:

1) cmts: total comment count 
2) AVcmts: average comments per post 
3) lks: total like count
4) AVlik: average likes per post
5) SMP: total number of posts (what if they have high engagement but few posts? Shoddy partner, maybe)
6) folr: total number of followers
7) foli: total number profile is following 
8) ALF: average number of likes per post per follower--almost 3-D** information here!!
9) ACF: average number of comments per post per follower--almost 3-D** information here, too!!

*>5, but some may be redundant
** we have the dimensions of number of likes or comments, 
'''
#new idea: fice will be populated from a string handle alone, and maybe have just a dictionary, 
#and then each value will be populated with a support function
#this way, a fice can be instaniated more easily, and the the data for a fice will be more organized and 
#more jasonizable
#another proposed scoring from what we saw online: total_num_likes + total_num_comments) / (num_followers * total_num_posts
class fice:
    def __init__(self, pfTag=''):
    
        self.name = pfTag #name for Kardder purposes
        self.pfTag = pfTag #tag for instascraping/loading purposes
        
        #creating Profile object from which to access data, using a passed string of the pf in question
        profile = Profile.from_username(loader.context, pfTag)

        #setting up data to be passed to member datafields
        num_followers = profile.followers
        num_followees = profile.followees
        total_num_likes = 0
        total_num_comments = 0
        total_num_posts = 0

        #grabbing the data, post-wise seems like only way atm...
        for post in profile.get_posts():
            total_num_likes += post.likes
            total_num_comments += post.comments
            total_num_posts += 1
        
        
        self.cmts = total_num_comments #total number of comments
        self.AVcmt = total_num_comments/total_num_posts #average comments per post
        
        self.lks = total_num_likes #total number of comments
        self.AVlik = total_num_likes/total_num_posts #average likes per post
        
        self.SMP = total_num_posts #social media prolificacy, total number of posts
        
        self.folr = num_followers #total number of followers
        self.foli = num_followees #total number following
        
        self.FFR = (num_followers/num_followees) #follower : following ratio, score of relative of raw "influence"
        
        #these will represent the relative engagement/impact a pf has, proportional to its followers
        
        self.ALF = (total_num_likes/total_num_posts)/num_followers #average likes : post : followers
        self.ACF = (total_num_comments/total_num_posts)/num_followers #average comments : posts : followers
```

```python
def print_fice_data(fice):
    print("Follower count: " + str(fice.folr), 
          "\nFollowee count: " + str(fice.foli), 
          "\nFFR: " + str(fice.FFR), 
          "\nTotal comments: " + str(fice.cmts), 
          "\nAverage comments per post: " + str(fice.AVcmt), 
          "\nTotal likes: " + str(fice.lks), 
          "\nAverage likes per post: " + str(fice.AVlik), 
          "\nSocial Media Prolificacy: " + str(fice.SMP), 
          "\nALF: " + str(fice.ALF), 
          "\nACF: " + str(fice.ACF))
```

```python
import instaloader
from datetime import datetime
from itertools import dropwhile, takewhile
import csv

class GetInstagramProfile():
    def __init__(self) -> None:
        self.L = instaloader.Instaloader()

    def download_users_profile_picture(self,username):
        self.L.download_profile(username, profile_pic_only=True)

    def download_users_posts_with_periods(self,username):
        posts = instaloader.Profile.from_username(self.L.context, username).get_posts()
        SINCE = datetime(2021, 8, 28)
        UNTIL = datetime(2021, 9, 30)

        for post in takewhile(lambda p: p.date > SINCE, dropwhile(lambda p: p.date > UNTIL, posts)):
            self.L.download_post(post, username)

    def download_hastag_posts(self, hashtag):
        for post in instaloader.Hashtag.from_name(self.L.context, hashtag).get_posts():
            self.L.download_post(post, target='#'+hashtag)

    def get_users_followers(self,user_name):
        '''Note: login required to get a profile's followers.'''
        self.L.login(input("input your username: "), input("input your password: ") ) 
        profile = instaloader.Profile.from_username(self.L.context, user_name)
        file = open("follower_names.txt","a+")
        for followee in profile.get_followers():
            username = followee.username
            file.write(username + "\n")
            print(username)
        

    def get_users_followings(self,user_name):
        '''Note: login required to get a profile's followings.'''
        self.L.login(input("input your username: "), input("input your password: ") ) 
        profile = instaloader.Profile.from_username(self.L.context, user_name)
        file = open("following_names.txt","a+")
        for followee in profile.get_followees():
            username = followee.username
            file.write(username + "\n")
            print(username)

    def get_post_comments(self,username):
        posts = instaloader.Profile.from_username(self.L.context, username).get_posts()
        for post in posts:
            for comment in post.get_comments():
                print("comment.id  : "+str(comment.id))
                print("comment.owner.username  : "+comment.owner.username)
                print("comment.text  : "+comment.text)
                print("comment.created_at_utc  : "+str(comment.created_at_utc))
                print("************************************************")

    def get_post_info_csv(self,username):
        with open(username+'.csv', 'w', newline='', encoding='utf-8') as file:
            writer = csv.writer(file)
            posts = instaloader.Profile.from_username(self.L.context, username).get_posts()
            for post in posts:
                print("post date: "+str(post.date))
                print("post profile: "+post.profile)
                print("post caption: "+post.caption)
                print("post location: "+str(post.location))
                
                posturl = "https://www.instagram.com/p/"+post.shortcode
                print("post url: "+posturl)
                writer.writerow(["post",post.mediaid, post.profile, post.caption, post.date, post.location, posturl,  post.typename, post.mediacount, post.caption_hashtags, post.caption_mentions, post.tagged_users, post.likes, post.comments,  post.title,  post.url ])
            
                for comment in post.get_comments():
                    writer.writerow(["comment",comment.id, comment.owner.username,comment.text,comment.created_at_utc])
                    print("comment username: "+comment.owner.username)
                    print("comment text: "+comment.text)
                    print("comment date : "+str(comment.created_at_utc))
                print("\n\n")


'''if __name__=="__main__":
    cls = GetInstagramProfile()
    #cls.download_users_profile_picture("best_gadgets_2030")
    #cls.download_users_posts_with_periods("best_gadgets_2030")
    #cls.download_hastag_posts("gadgets")
    #cls.get_users_followers("best_gadgets_2030")
    #cls.get_users_followings("best_gadgets_2030")
    #cls.get_post_comments("laydline")
    cls.get_post_info_csv("alexjory")'''
    
cls = GetInstagramProfile()
```
